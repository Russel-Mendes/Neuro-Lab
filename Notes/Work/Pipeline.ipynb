```python
#imports 
import czifile
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from PIL import Image
```

```python
PATH_TO_DATA = "D:\\Program Projects\\Lab Project\\Data\\CZ Files\\Neurons\\"
```

```python
def get_files(path_):
    files = glob.glob(path_ + '/**/*.czi', recursive=True)
    for file in files:
        yield file
```

```python
def load_czi(path):
    czi = czifile.imread(path)
    shape = czi.shape
    return czi, shape
```

```python
def save_path(foward_slice, current_path):
    split = current_path.split("\\")
    name = split[-1].split(".")[0]
    split = split[foward_slice:len(split)-2]
    separator = '\\'
    path = separator.join(split)
    try: 
        os.makedirs(path)
    except:
        pass
    return path +  "\\"+ name
```

```python
def extract_img(czi, h_w, filter_, slides_, root):
    img = czi[0,:,:,0,:,:,0]
    for i in range(slides_):
        for j in range(filter_):
            name_tag = "_Slide_%i_Filter_%i"% (i, j)
            sel_img = img[i,j,:,:]
            sel_img = 255.0/np.max(sel_img)*sel_img
            name = root+ name_tag
            
            tiff_img = sel_img
            tiff_img = Image.fromarray(sel_img)
            tiff_img.save(name+".tiff", "tiff")
            
            sel_img = sel_img.astype(dtype=np.uint8)
            int_img = Image.fromarray(sel_img)
            int_img.save(name+".png", "png")
```

```python
my_gen = get_files(PATH_TO_DATA)
counter = 0
foward_slice = 4
for path in my_gen:
    counter = counter + 1
    czi,channel = load_czi(path)
    h_w = channel[4]
    filter_ = channel[2]
    slides_ = channel[1]
    s_p = save_path(foward_slice, path)
    extract_img(czi, h_w, filter_, slides_, s_p)
    if counter == 100:
        break
```
