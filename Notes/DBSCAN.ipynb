## DBSCAN

DBSCAN stands for Density-Based Spatial Clustering of Applications with Noise. This is an unsupervised learning algorithm that seeks to cluster data points. It main metrics are based on the density of the data points. It starts at an arbitrary points in an n-dimensional space and clusters with a radius denoted as epsilon. The program continues until it reaches the threshold required to declare a cluster. If a point does not fit within the density cluster it declares it as noise and ignores it. One limitation of this model is that it struggles with varying densities of point. 
A reference source can be found [here](https://medium.com/@elutins/dbscan-what-is-it-when-to-use-it-how-to-use-it-8bd506293818). Another good source is [here](https://towardsdatascience.com/dbscan-clustering-for-data-shapes-k-means-cant-handle-well-in-python-6be89af4e6ea)

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
#Creating the Data Set

#4 distinct clusters
x0 = -2 * np.random.rand(100,2)
x1 =  2 * np.random.rand(100,2)
x2 = -1 * np.random.rand(100,2)
x3 =1 + 3 * np.random.rand(100,2)

#offsetting
x0 = np.array([[i[0]-5, i[1]-5] for i in x0])
x1 = np.array([[i[0], i[1]] for i in x1])
x2 = np.array([[i[0]-5, i[1]+5] for i in x2])
x3 = np.array([[i[0]+3, i[1]+5] for i in x3])

#x[100:200, :] = x0
```

```python
data_whole = np.zeros((400,2))
data_whole[0:100, :] = x0
data_whole[100:200, :] = x1
data_whole[200:300, :] = x2
data_whole[300:400, :] = x3
```

```python
plt.scatter(data_whole[ : , 0], data_whole[ :, 1], s = 25, color='r')
plt.grid()
#Note
#The lower left cluster is x0
#The center right cluster is x1
#The tight upper left cluster is x2
#The loose upper right cluster is x3
```

```python
# Model setup
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_whole)
dbscan = DBSCAN(eps=0.123, min_samples = 2)
clusters = dbscan.fit_predict(data_scaled)
```

```python
# plot the cluster assignments
plt.scatter(data_whole[:, 0], data_whole[:, 1], c=clusters, cmap="plasma")
plt.show()
```
