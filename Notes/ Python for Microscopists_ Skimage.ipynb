```python
# Edge Detection
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters import roberts, sobel, prewitt, scharr
img_data = io.imread('D:\Program Projects\Lab Project\input\convert.png', as_gray = True)
plt.imshow(img_data, cmap ='copper')
```

```python
plt.imshow(roberts(img_data), cmap = 'copper')
```

```python
plt.imshow(sobel(img_data), cmap = 'copper')
```

```python
plt.imshow(prewitt(img_data), cmap = 'copper')
```

```python
plt.imshow(scharr(img_data), cmap = 'copper')
```

```python
from skimage.feature import canny
plt.imshow(canny(img_data, sigma = .1), cmap = 'copper')
```

```python
#Image deconvolutions
from skimage import restoration
import numpy as np

psf = np.ones((3,3))/9

decon_image,_ = restoration.unsupervised_wiener(img_data, psf)

plt.imshow(decon_image, cmap = 'copper')
```
