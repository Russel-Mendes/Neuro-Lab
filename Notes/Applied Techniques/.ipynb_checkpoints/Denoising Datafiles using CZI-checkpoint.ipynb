```python
import czifile
import matplotlib.pyplot as plt
czi = czifile.imread("data\RT_HC_10X_12-4-19_PLMD91_A.czi")
img = czi[0,0,0,:,0,:,:,0]
img = img[0,:,:]
plt.figure(figsize = (20,20))
plt.title("Original")
plt.imshow(img,cmap = "hot")
```

```python
from scipy import ndimage as nd
gaus_img = nd.gaussian_filter(img, sigma=10)
plt.figure(figsize = (20,20))
plt.title("Gaussian")
plt.imshow(img,cmap = "hot")
```

```python
median_img = nd.median_filter(img, size=10)
plt.figure(figsize = (20,20))
plt.title("Median")
plt.imshow(img,cmap = "hot")
```
